## Task 04 – Fake News Detection using NLP

This project focuses on identifying whether a news article is real or fake using Natural Language Processing techniques and machine learning classification.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fake_news_dataset.csv")
df.head()
```

```python
X = df["text"]
y = df["label"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words="english", max_df=0.7)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Conclusion

The model successfully classified news articles as real or fake using NLP and logistic regression. TF-IDF helped convert text into numerical form, allowing the algorithm to detect patterns in language usage.
